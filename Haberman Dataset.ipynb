# Univariate Analysis

```python
status_1 = cancer_data.loc[cancer_data['status']==1]
status_2 = cancer_data.loc[cancer_data['status']==2]
label=(1,2)
plt.plot(status_1['nodes'],np.zeros_like(status_1['nodes']),'o')
plt.plot(status_2['nodes'],np.zeros_like(status_2['nodes']),'o')
plt.xlabel('Nodes')
plt.legend(labels=label)
plt.show()
```

```python
sns.FacetGrid(cancer_data,hue='status',height=5).map(sns.distplot,'nodes').add_legend()
plt.show()
```

## Observation
    The Probability Density Function(PDF) of both Status 1 & 2 are overlapping each other while plotting for Nodes Feature.
    The peak of Status 2 is inside the peak of Status 1 with a little wide spread of values above 5 nodes.
    Person with nodes less than 5 had both higher and low chances of survival.

```python
sns.FacetGrid(cancer_data,hue='status',height=5).map(sns.distplot,'age').add_legend()
plt.show()
```

## Observation
    We couldnt predict the Survival Status of Cancer Patient based on the Age Feature as both the PDFs are highly overlapping each other.

```python
sns.FacetGrid(cancer_data,hue='status',height=5).map(sns.distplot,'year').add_legend()
plt.show()
```

## Observation
    Similar overlapping of PDFs is observed with the Year Feature.
    Hence, Nodes is slightly better to compare the status than the Age and Year Features

# PDF & CDF

```python
counts,bin_edges= np.histogram(status_1['nodes'],bins=10,density=True)
print(counts)
print(bin_edges)
```

```python
pdf=counts/sum(counts)
print('PDF')
print(pdf)
print('\n')


cdf=np.cumsum(pdf)
print('CDF')
print(cdf)
print('\n')

print('Bin Edges Shape')
print(bin_edges.shape)
print('\n')

print('PDF Shape')
print(pdf.shape)

labels=('PDF','CDF')
plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:],cdf)
plt.legend(labels=labels)
plt.show()
```

## Observation
    There is a 85% survival rate if the nodes count >= 5 and the survival rate is almost 100% if nodes count >= 30

```python
counts,binedges = np.histogram(status_1['nodes'],bins=10,density=True)
counts1,binedges1 = np.histogram(status_2['nodes'],bins=10,density=True)

pdf_1 = counts/sum(counts)
cdf_1 = np.cumsum(pdf_1)

pdf_2 = counts1/sum(counts1)
cdf_2 = np.cumsum(pdf_2)

labels=('pdf_1','pdf_2','cdf_1','cdf_2')
plt.plot(binedges[1:],pdf_1)
plt.plot(binedges1[1:],pdf_2)
plt.plot(binedges[1:],cdf_1)
plt.plot(binedges1[1:],cdf_2)
plt.legend(labels=labels)
plt.show()
```

## Observation
    Patients with nodes >=5 also have approx 55% less Survival rate
